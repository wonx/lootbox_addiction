Converting `parsedresults.csv` dataset to the same format used in the project so it can be parsed and split into daily csv files.

```python
import pandas as pd
```

```python
df = pd.read_csv('parsedresults.csv')
display(df.head())
```

Drop unnecessary columns

```python
df.drop(columns=['batch', 'number', 'latest', 'index', 'seconds', 'window.low', 'window.high', 'identity'], inplace=True)
display(df.head())
```

Add datetime column at the beginning, converting the unix timestamp to a date time object at the UTC timezone.

```python
import datetime
df['datetime'] = pd.to_datetime(df['timestamp'], unit='s', utc=True).dt.strftime('%Y-%m-%d %H:%M:%S')
df = df[['datetime', 'timestamp', 'user', 'src', 'out', 'time']]
display(df.head())
```

```python
df.to_csv('parsedresults_harmonized.csv', index=False)
```

Now the harmonized dataset `parsedresults_harmonized.csv` is ready to be split to daily csv files using the `split_csv_bydate.py` script in the `data_ingestion` folder.
